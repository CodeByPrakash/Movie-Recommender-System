# now its a list,i want to convert the entire genres into this above specific format

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

# now we need to do same format for keywords

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'].iloc[0]
```

```python
# we need first 3 original name of cast
```

```python
def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
          L.append(i['name'])
          counter+=1
        else:
            break
    return L  
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

# from crew ,we find directors only

```python
def find_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
          L.append(i['name'])
          break
    return L  
```

```python
movies['crew']=movies['crew'].apply(find_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
ps.stem('activities')
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
def recommander(movie):
    movie_index=new_movies[new_movies['title']==movie].index[0]
    distances=similarity_matrix[movie_index]
    recommanded_movie_list=sorted(list(enumerate((distances))),reverse=True,key=lambda x:x[1])[1:11]

    for i in recommanded_movie_list:
        print(new_movies.iloc[i[0]]['title'])
        
```
